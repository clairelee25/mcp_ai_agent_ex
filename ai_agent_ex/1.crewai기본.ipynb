# CrewAI ê¸°ë³¸ ì‹¤ìŠµ

```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¡œë”©
from crewai import Agent, Task, Crew
from langchain_openai import ChatOpenAI
```

## ì—ì´ì „íŠ¸ ê°ì²´ ìƒì„±
### ëª©ì°¨ ì‘ì„± Agnet ì •ì˜

```python
# ëª©ì°¨ ì„¤ì • ì—ì´ì „íŠ¸
outline_generator = Agent(
    role='outline_generator',
    # ì£¼ì–´ì§„ ì£¼ì œì— ëŒ€í•œ êµ¬ì¡°í™”ëœ ëª©ì°¨ë¥¼ ìƒì„±í•˜ëŠ” ì—ì´ì „íŠ¸ì…ë‹ˆë‹¤. í•œêµ­ì–´ë¡œ ë‹µë³€í•©ë‹ˆë‹¤.
    goal="Create structured outlines for articles on given topics. answer in Korean.",
    # ê°€ë²¼ìš´ ì‘ì—…ì„ ìˆ˜í–‰í•˜ëŠ” ì—ì´ì „íŠ¸ì…ë‹ˆë‹¤.
    llm=ChatOpenAI(
        model="gpt-4o-mini",
        temperature=0.0,
        max_tokens=1000,
    ),
    # ëª©ì°¨ì„¤ì •ì¹˜ ë§¥ë½ ì„¤ì •, ë„ˆëŠ” ë‹¤ì–‘í•œ ì£¼ì œì— ëŒ€í•´ í¬ê´„ì ì¸ ëª©ì°¨ë¥¼ ì‘ì„±í•˜ëŠ” ì „ë¬¸ê°€ì…ë‹ˆë‹¤.  
    backstory="""
        You are an expert at organizing information and creating comprehensive outlines for various subjects.
        """
)
```

### ë³¸ë¬¸ ì‘ì„± Agent ì •ì˜

```python
# ë³¸ë¬¸ ì‘ì„± ì—ì´ì „íŠ¸
writer=Agent(
    role='writer',
    goal="Create engagin content based on reshearch. answer in Korean.",
    llm=ChatOpenAI(
        model="gpt-4o-mini",
        max_tokens=3000,
    ),
    backstory="""
    You are a skilled writer who can transform complex information into readable content.
    """
)
```

## Task ì •ì˜
### ëª©ì°¨ task

```python
# ëª©ì°¨ ë§Œë“œëŠ” task
outline_task=Task(
    name="outline_task",
    description="Create a detailed outline for the article about AI\'s impact on job markets.",
    agent=outline_generator,
    # ì£¼ì œëŠ” í•œêµ­ì–´ë¡œ ì‘ì„±í•©ë‹ˆë‹¤.
    expected_output="""
    A comprehensive outline convering the main aspects of AI\'s influence on employment
    """,
)
```

### ë³¸ë¬¸ task

```python
# ë³¸ë¬¸ ì‘ì„± task
writing_task = Task(
    name="article_Task",
    description="Write a detailed article about the findings from the research",
    agent=writer,
    # ë³¸ë¬¸ì€ í•œêµ­ì–´ë¡œ ì‘ì„±í•©ë‹ˆë‹¤.
    expected_output="""
    An engaging article disussing AI's influence on job markets.
    """
)
```

# Crew ì •ì˜
- agentì™€ taskë¥¼ ë¬¶ëŠ” ì—­í• 

```python
# crew ì •ì˜
ai_impact_crew=Crew(
    agents=[outline_generator, writer],
    tasks=[outline_task, writing_task],
    verbose=True
)
```
